```python
plt.hist(np.random.rand(100000), density=True, bins=150, histtype="step", color="blue", label="rand")
plt.hist(np.random.randn(100000), density=True, bins=150, histtype="step", color="red", label="randn")

plt.axis([-4, 4, 0, 1.2])
plt.legend(loc = "upper left")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

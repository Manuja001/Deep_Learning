```python
uniform = np.random.rand(100000)
normal = np.random.randn(100000)
exponential = np.random.exponential(scale=1.0, size=100000)

plt.hist(uniform, density=True, bins=100, histtype="step", color="blue", label="Uniform")
plt.hist(normal, density=True, bins=100, histtype="step", color="green", label="Normal")
plt.hist(exponential, density=True, bins=100, histtype="step", color="red", label="Exponential")

plt.axis([-2.5, 7, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2


fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')


ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title(r'$Z = X^2 + Y^2$')

plt.show()
```
